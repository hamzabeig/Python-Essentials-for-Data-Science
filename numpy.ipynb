#linear algebra

```python
#inverse of matrix
arr1 = np.array([[2,3],[7,8]])
arr1
np.linalg.inv(arr1)
```

```python
#eigenvalues and eigenvectors
arr1 = np.array([[2,3],[7,8]])
arr1
eigvalues, eigvectors = np.linalg.eig(arr1)
eigvectors
```

```python
#diagnolization
# A = PDP-1
# P = [eigenvectors]
# D = [all values ARE ZERO, diagnol eigen values]
# P-1 = P-1

A = np.array([[2,3],[7,8]])
eigenval, eigvec = np.linalg.eig(A)
P  = eigvec
P_inv = np.linalg.inv(P)
D = np.zeros((2,2))
for i in range(2):
  D[i,i] = eigenval[i]
D
# A = PDP-1
P @ D @ P_inv
```
